## Day 1

How quickly is the sub descending?



```python
import sol as sol
sol.Day1().sol()

```

Lets see this as a descending graph


```python
import pandas as pd
depths = sol._get_lines(day=1, parser=int)
neg_d = [-x for x in depths]
df2 = pd.DataFrame(dict(depths=neg_d, count=[x for x in range(len(depths))]))

import altair as alt
chart = alt.Chart(df2)
chart.mark_point().encode(x='count', y='depths')
```

## Day 2

Calculate the subs position after following stream of movement instructions.

The only instructions seem to be:
- forward [int]
- down [int]
- up [int]


```python
day2 = sol.Day2()
day2.part1()
```

Let's see how the sub is moving then, let's see it's descend path


```python
d2df = pd.DataFrame(dict(horz=[x[0] for x in day2.positions], depth=[x[1] for x in day2.positions]))
chart2 = alt.Chart(d2df)
chart2.mark_point().encode(x='horz', y='depth')
```

Part 2, introduces a third value 'aim' that starts are 0.
- down X increases your aim by X units.
- up X decreases your aim by X units.
- forward X does two things:
  - increases your horizontal position by X units.
  - increases your depth by your aim multiplied by X.

Let's see how what that ends up with, and how it graphs

```python
day2.part2()
d2df = pd.DataFrame(dict(horz=[x[0] for x in day2.positions], depth=[x[1] for x in day2.positions]))
chart2 = alt.Chart(d2df)
chart2.mark_point().encode(x='horz', y='depth')
```

Final results are:
